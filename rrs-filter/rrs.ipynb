# Implementation of a Recursive Running Sum (RRS) Filter and visualization

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
# time parameters
fs = 1e5
T = 1
num_samples = int(fs * T)
time_arr = np.linspace(0.,T, num_samples, endpoint=False)

# sinusoid signal
f1 = 250
sinusoid = np.sin(2*np.pi*f1*time_arr)
#noise
power = 1e-5
noise = np.sqrt(power) * np.random.randn(num_samples)

x = sinusoid + noise
```

```python
%matplotlib
#plots
fig, ax = plt.subplots()
ax.plot(time_arr, x, lw=0.5)
plt.show()
```
